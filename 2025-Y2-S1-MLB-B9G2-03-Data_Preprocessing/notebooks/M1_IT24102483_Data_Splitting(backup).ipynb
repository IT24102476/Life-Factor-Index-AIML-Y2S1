```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
import pandas as pd

# Load or create your DataFrame (replace with your actual data)
# Example: df = pd.read_csv('your_dataset.csv')
# Sample DataFrame for demonstration
df = pd.DataFrame({
    'Education Level': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor',
                       'High School', 'Master', 'Bachelor', 'PhD', 'Master'] * 10
})

# Set seaborn style
sns.set_style("whitegrid")

print("\n--- Generating EDA Visualizations... ---")

# Create count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Education Level', data=df, palette='viridis')
plt.title('Count of Individuals by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=30)  # Rotate labels if too long
plt.tight_layout()
plt.show()

print("--- EDA visualizations completed successfully! ---")

# KFold Cross Validation
n_splits = 6
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

print(f"\n--- Splitting data into {n_splits} folds for cross-validation ---")
print("This process will create 6 pairs of training and testing datasets.")

for i, (train_index, test_index) in enumerate(kf.split(df)):
    print(f"\nFold {i+1}:")
    print(f"  Training set size: {len(train_index)} rows")
    print(f"  Testing set size: {len(test_index)} rows")

print("\n--- Data splitting process completed successfully. ---")
print("You now have the indices to create your training and testing sets for each fold.")
```
